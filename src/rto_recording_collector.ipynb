```python
import os
import pandas as pd
import shutil
from pathlib import Path
```

```python
cwd = os.getcwd()
output = str(Path(cwd).parents[0]) + "\\Today\\Recording Proof"
excel_file = pd.read_excel("#RTO Recording AWB.xlsx")
or_folder = str(Path(cwd).parents[0]) + "\\(04) NDR Call Recordings"
```

```python
awb_list = excel_file['AWB'].astype(str).tolist()
found_files = []
for awb in awb_list:
    for file_name in os.listdir(or_folder):
        if awb in file_name: 
            found_files.append(file_name)
```

```python
for f in found_files:
    if f.lower().endswith((".mp3")):
        shutil.copy(os.path.join(or_folder, f), os.path.join(output, f))
```
